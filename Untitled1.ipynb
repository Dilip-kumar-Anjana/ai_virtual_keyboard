```python
import cv2
from cvzone.HandTrackingModule import HandDetector

cap = cv2.VideoCapture(0)
cap.set(3, 1280)
cap.set(4, 720)
detector = HandDetector(detectionCon=0.8)

if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()
    
class Button():
    def __init__(self, pos, text, size=[85, 85]):
        self.pos = pos
        self.size = size
        self.text = text
        
        cv2.rectangle(img, self.pos, self.size, (175, 0, 175), cv2.FILLED)
        cv2.putText(img, self.text, (self.pos[0]+25, self.pos[1]+25),
                cv2.FONT_HERSHEY_PLAIN, 5, (255, 255, 255), 5)
    


while True:
    success, img = cap.read()
    if not success:
        print("Failed to capture image")
        break
    hands, img = detector.findHands(img)  # Updated method
    


    if hands:
        # Hand 1
        hand1 = hands[0]
        lmList1 = hand1["lmList"]  # List of 21 Landmark points
        bbox1 = hand1["bbox"]  # Bounding box info x,y,w,h
        centerPoint1 = hand1['center']  # center of the hand cx,cy
        handType1 = hand1["type"]  # Handtype Left or Right

        # Print landmarks and bounding box info
        print(lmList1, bbox1, centerPoint1, handType1)
        
    mybutton = Button([100,100],"Q")

    cv2.imshow("Image", img)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
